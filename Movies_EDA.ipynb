```python
import pandas as pd
data = pd.read_csv("/content/tmdb_5000_movies.csv")
data.head()
data.shape
data.columns
data.info()
data.describe()
data.isnull().sum()
data['overview'] = data['overview'].fillna("Unknown")
data['vote_average'] = data['vote_average'].fillna(data['vote_average'].mean())
data = data.drop(columns=['homepage'])
data.isnull().sum()
import matplotlib.pyplot as plt
import seaborn as sns

##  Distribution of Movie Popularity
"""
This visualization shows how movie popularity is distributed across the dataset.
Most movies have low popularity, while only a few movies achieve very high popularity, indicating a right-skewed distribution.
"""
plt.figure(figsize=(8, 6))
plt.hist(data['popularity'], bins=30)
plt.xlabel("Popularity")
plt.ylabel("Number of Movies")
plt.title("Distribution of Movie Popularity")
plt.show()

##  Top 10 Most Popular Movies
"""
This bar chart highlights the top 10 movies based on popularity scores.
These movies represent blockbuster films that received high audience attention and engagement.
"""
top_movies = data.sort_values(by='popularity', ascending=False).head(10)
plt.figure(figsize=(8, 6))
plt.barh(top_movies['original_title'], top_movies['popularity'])
plt.xlabel("Popularity")
plt.ylabel("Movie Title")
plt.title("Top 10 Most Popular Movies")
plt.show()

##  Distribution of Movie Ratings
"""
This graph shows how average movie ratings are distributed.
Most movies fall within the 5 to 7 rating range, indicating moderate audience satisfaction.
"""
plt.figure(figsize=(8, 6))
plt.hist(data['vote_average'], bins=10)
plt.xlabel("Average Rating")
plt.ylabel("Number of Movies")
plt.title("Distribution of Movie Ratings")
plt.show()

##  Votes vs Average Ratings
"""
This scatter plot illustrates the relationship between the number of votes and average ratings.
Movies with higher vote counts tend to have more stable and reliable ratings compared to movies with fewer votes.
"""
plt.figure(figsize=(8, 6))
plt.scatter(data['vote_count'], data['vote_average'])
plt.xlabel("Vote Count")
plt.ylabel("Average Rating")
plt.title("Votes vs Ratings")
plt.show()

##  Key Insights & Conclusion
"""
- Most movies have low popularity, while only a few blockbuster movies dominate audience attention.
- The majority of movies receive average ratings between 5 and 7, indicating moderate viewer satisfaction.
- Movies with a higher number of votes tend to have more consistent and trustworthy ratings.

Overall, this exploratory data analysis helped uncover important trends and patterns in movie popularity and audience preferences.
"""


```
